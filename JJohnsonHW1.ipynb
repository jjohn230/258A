```julia
using Convex
using ECOS
```

```julia
using PyPlot
```

```julia
n = 100
A = randn(n,n)
b = randn(n,1)
x1 = Variable(n)
l = 5
p = minimize(sum_squares(A*x1-b)+l*norm(x1,1))
solve!(p, ECOSSolver())
```

```julia
figure(figsize=(20,5))
r1 = -1:0.1/50:1
plt[:hist](x1.value,50,facecolor = "w")
plot(r1,abs(r1),"k")
axis([-1,1,0,7])
```

```julia
n = 100
A = randn(n,n)
b = randn(n,1)
x2 = Variable(n)
l = 5
p = minimize(sum_squares(A*x2-b)+l*norm(x2,2))
solve!(p, ECOSSolver())
```

```julia
figure(figsize=(20,5))
r2 = -1:0.1/50:1
plt[:hist](x2.value,50,facecolor = "w")
plot(r2,(r2).^2,"k")
axis([-1,1,0,7])
```

```julia
n = 100
A = randn(n,n)
b = randn(n,1)
x3 = Variable(n)
l = 1/10
p = minimize(sum_squares(A*x3-b)+l*(max(norm(x3,1)-0.5,0)))
solve!(p, ECOSSolver())
```

```julia
figure(figsize=(20,5))
r3 = -1:0.1/50:1
plt[:hist](x3.value,50,facecolor = "w")
plot(r3,max(abs(r3)-0.5,0),"k")
axis([-1,1,0,7])
```

```julia
n = 100
A = randn(n,n)
b = randn(n,1)
x = Variable(n)
p = minimize(sum_squares(A*x-b)-sum(log(-x+1))-sum(log(x+1)))
solve!(p, ECOSSolver())
```

```julia
figure(figsize=(20,5))
r = -1:(1/50):1
plt[:hist](x.value,50,facecolor = "w")
plot(r,-log(r+1)-log(-r+1),"k")
axis([-1,1,0,10])
```
